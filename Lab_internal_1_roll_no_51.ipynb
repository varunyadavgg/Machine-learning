```python
#  1
#incase booking
class Train:
      def __init__(self,nt1,nt2,nt3):
        self.nt1=nt1
        self.nt2=nt2
        self.nt3=nt3
class Reservation(Train):
  def __init__(self,nt1,nt2,nt3,bt1,bt2,bt3):
    Train.__init__(self,nt1,nt2,nt3)
    self.bt1=bt1
    self.bt2=bt2
    self.bt3=bt3 
  def display(self):
    print("total seats present in tier_1:",self.nt1)
    print("total seats present in tier_2:",self.nt2)
    print("total seats present in tier_3:",self.nt3)
    print('seats booked in Tier_1:',self.bt1)
    print('seats booked in Tier_2:',self.bt2)
    print('seats booked in Tier_3:',self.bt3)
  def Book_new(self):
    self.at1=self.nt1-self.bt1
    self.at2=self.nt2-self.bt2
    self.at3=self.nt3-self.bt3
    k=int(input("enter your tier list: "))
    if k ==1:
          if self.at1>0:
                print("Your Seat is booked !!")
                self.at1-=1
                self.bt1+=1
          else:
                print("all tickets are booked!!!")
    elif k ==2:
          if self.at2>0:
                print("Your Seat is booked !!")
                self.at2-=1
                self.bt2+=1
          else:
                print("all tickets are booked!!!")
    elif k ==3:
          if self.at3>0:
                print("Your Seat is booked !!")
                self.at3-=1
                self.bt3+=1
          else:
                print("all tickets are booked!!!")
    else:
          print("Wrong tier List !!!")
  def Cancel(self):
    self.at1=self.nt1-self.bt1
    self.at2=self.nt2-self.bt2
    self.at3=self.nt3-self.bt3
    p=int(input("enter your tier list: "))
    if p ==1:
          if self.at1<self.nt1:
                print("Your Seat is Cancelled !!")
                self.at1+=1
                self.bt1-=1
          else:
                print("You dont own a ticket!!!")
    elif p ==2:
          if self.at2<self.nt2:
                print("Your Seat is Cancelled !!")
                self.at2+=1
                self.bt2-=1
          else:
                print("You dont own a ticket!!!")
    elif p ==3:
          if self.at3<self.nt3:
                print("Your Seat is Cancelled !!")
                self.at3+=1
                self.bt3-=1
          else:
                print("You dont own a ticket!!!")
    else:
          print("Wrong tier List !!!")
  def status(self):
        print("*******Status*****")
        print("Tier 1 : Available = ",self.at1," ,Booked = ",self.bt1," ,Total seats  = ",self.nt1)
        print("Tier 2 : Available = ",self.at2," ,Booked = ",self.bt2," ,Total seats  = ",self.nt2)
        print("Tier 3 : Available = ",self.at3," ,Booked = ",self.bt3," ,Total seats  = ",self.nt3)
        
nt1=int(input("Enter total number of seats in tier 1: "))
nt2=int(input("Enter total number of seats in tier 2: "))
nt3=int(input("Enter total number of seats in tier 3: "))

bt1=int(input("Enter total number of Booked seats in tier 1: "))
bt2=int(input("Enter total number of Booked seats in tier 2: "))
bt3=int(input("Enter total number of Booked seats in tier 3: "))


A=Reservation(nt1,nt2,nt3,bt1,bt2,bt3)

q=input("Enter the action book or cancel : ")
if q =="book" or q == "Book":
  A.Book_new()
  print ("\n\ncurrent booking status is : \n")
  A.status()
elif q =="cancel" or q == "Cancel":
  A.Cancel()
  print ("\n\ncurrent booking status is : \n")
  A.status()
else:
  print("Wrong action input!!")


```
